```
# Drop ["date", "lights"] features
energy_dataset = energy_dataset.drop(columns = ['date', 'lights'])
energy_dataset.head(10)
```

```
# create a dictionary of current and new dataset column names
new_name = {'T1':'Temp1_kitchen', 'RH_1':'Humid1_kitchen', 'T2':'Temp2_living_room', 'RH_2':'Humid2_living_room', 'T3':'Temp3_laundry_room', 'RH_3':'Humid3_laundry_room', 'T4':'Temp4_office_room', 'RH_4':'Humid4_office_room', 'T5':'Temp5_bathroom', 'RH_5':'Humid5_bathroom', 'T6':'Temp6_outside', 'RH_6':'Humid6_outside', 'T7':'Temp7_ironing_room', 'RH_7':'Humid7_ironing_room', 'T8':'Temp8_teenage_room2', 'RH_8':'Humid8_teenage_room2', 'T9':'Temp9_parents_room', 'RH_9':'Humid9_parents_room',
            'T_out':'Temp_from_wheather_station', 'Press_mm_hg':'pressure_from_wheather_station', 'RH_out':'Humid_from_wheather_station', 'Wind speed':'wind_speed_from_wheather_station', 'Visibility':'Visibility_from_wheather_station', 'Tdewpoint':'Dewpoint_from_wheather_station', 'rv1':'Random_variable1', 'rv2':'Random_variable2'
           }
            
```

```
# Rename features to make each descriptive
wheath_report = energy_dataset.rename(columns=new_name)
wheath_report.head()
```

```
# Normalization of features
from sklearn.preprocessing import MinMaxScaler

scaled = MinMaxScaler()
norm_features = pd.DataFrame(scaled.fit_transform(wheath_report), columns=wheath_report.columns)

X_features = wheath_report.drop(columns=['Appliances'])
y_features = wheath_report['Appliances']


```

```
# Split dataset into training and validation dataset in the ratio .7:.3
from sklearn.model_selection import train_test_split
X_train, X_validate, y_train, y_validate = train_test_split(X_features, y_features, test_size=.3, random_state=42)
```

```
# Define and fit model
from sklearn.linear_model import LinearRegression

LNR_Model = LinearRegression().fit(X_train, y_train)
```

```
# plot a regression graph to visualize the relationship
#import seaborn as sns
#import matplotlib.pyplot as plt
#%matplotlib inline

#sns.regplot(x=X_train, y=, data=LNR_Model)
```

```
# Penalization of the features
from sklearn.linear_model import Lasso, Ridge
regression_lasso = Lasso(alpha=0.001)
regression_lasso.fit(X_train, y_train)

regression_ridge = Ridge(alpha = 0.5)
regression_ridge.fit(X_train, y_train)



```

```
# Effects of regularization
def get_weights_data(model, feat, col_name):
  # function that returns the weight of each feature
  weight = pd.Series(model.coef_, feat.columns).sort_values()
  weights_data = pd.DataFrame(weight).reset_index()
  weights_data.columns = ['Features', col_name]
  weights_data[col_name].round(3)
  return weights_data

LNR_Model_var = get_weights_data(LNR_Model, X_train, 'LNR_Model\'s_Weight')
Ridge_weight_var = get_weights_data(regression_ridge, X_train, 'Weight for Ridge Method')
Lasso_weight_var = get_weights_data(regression_lasso, X_train, 'Weight for Lasso Method')


output_weights = pd.merge(LNR_Model_var, Ridge_weight_var, on='Features')
output_weights = pd.merge(output_weights, Lasso_weight_var, on='Features')
pd.set_option('display.max_columns', None)
output_weights.head(20)

  
```

```
# check coefficients of determination of R squared, intercept and slope
R_sq = LNR_Model.score(X_train, y_train)
print("Coefficient of determination is: ", R_sq)

print("Intercept is: ", LNR_Model.intercept_)
print("The slope is: ", LNR_Model.coef_)

```

```
# Model Prediction
y_predicted = LNR_Model.predict(X_validate)
y_predicted
```

```
# Evaluate the performance of the model using Mean Absolute Error
from sklearn.metrics import mean_absolute_error
 
MAE_Err = mean_absolute_error(y_validate, y_predicted)
MAE_Err
```

```
# Evaluate performance using residual sum of squares (RSS)
import numpy as np
RSS = np.sum(np.square(y_validate - y_predicted))
print(round(RSS, 3))
```

```
# Evaluate performance using R-Squared
from sklearn.metrics import r2_score
rScore = r2_score(y_validate, y_predicted)
round(rScore,3)
```

```
# Evaluate performance using Root mean square error
from sklearn.metrics import mean_squared_error 
RMSE = np.sqrt(mean_squared_error(y_validate, y_predicted))
round(RMSE,3)
```
